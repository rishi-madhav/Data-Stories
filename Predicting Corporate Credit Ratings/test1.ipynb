```python
import pandas as pd
import numpy as np
```

```python
corp_data = pd.read_csv('corporateCreditRatingWithFinancialRatios.csv')
corp_data.shape
```

```python
sic_code = pd.read_csv('sicdata.csv')
sic_code.shape
```

```python
for index in corp_data.index:
    to_replace = sic_code[sic_code['SIC Code'] == corp_data.loc[index, 'SIC Code']]
    for i in to_replace['Industry Title'].index:
        corp_data.loc[index, 'Sector'] = to_replace['Industry Title'][i]

corp_data.shape
```
